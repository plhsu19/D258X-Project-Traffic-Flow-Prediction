```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```
# linear training datasets
X = 15 + 5* np.random.rand(60, 1)
Y = -2 * X + 40 + np.random.randn(60, 1)
print(X)
print(Y)
```

```
# linear regression predictions
lrmodel = LinearRegression()
lrmodel.fit(X, Y)
Y_predict = lrmodel.predict(X)
```

```
plt.style.use('default')
plt.scatter(X, Y, label='training dataset')
plt.plot(X, Y_predict, label='linear regression predictions', color='green', linewidth=2.0)
plt.xlabel('x1', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.legend(fontsize=12)
plt.show()
```
