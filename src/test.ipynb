```

from matplotlib import pyplot as plt
x = list(range(1, 11))
y = [1, 2, 1, 3, 5, 6, 8, 2, 1, 9]

for i in range(len(y)):
  y[i] = y[i] * 0.1

plt.figure(figsize=(5, 10))
ax = plt.subplot()
plt.plot(x, y)
ax.set_xticks([2, 4, 10])
ax.set_xticklabels(['Feb', 'Apr', 'Oct'])
ax.set_yticks([0.2, 0.4, 0.9])
ax.set_yticklabels(['20%', '40%', '90%'])
plt.show()


```
